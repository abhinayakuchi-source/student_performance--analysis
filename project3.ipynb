```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'Student':['A','B','C','D','E'],
    'Maths':[80,65,90,75,85],
    'Science':[85,70,88,72,90],
    'English':[78,60,92,80,84]
}

df = pd.DataFrame(data)

df
```

```python
print(df.info())
print(df.describe())
```

```python
df['Average'] = df[['Maths','Science','English']].mean(axis=1)

print(df)
```

```python
plt.bar(df['Student'], df['Average'])
plt.title("Average Marks of Students")
plt.xlabel("Students")
plt.ylabel("Average Marks")
plt.show()
```

```python
subject_avg = df[['Maths','Science','English']].mean()

print(subject_avg)
```

```python
subject_avg.plot.pie(autopct='%1.1f%%')
plt.title("Subject Contribution")
plt.ylabel("")
plt.show()
```
